```python
import imageio
import matplotlib.pyplot as plt
from code.fourier_filter import fourier_filter

img = imageio.imread('data/screening_mammogram.png')
img_filtered = fourier_filter(img, True, 10)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
axes[0].imshow(img, cmap='gray')
axes[1].imshow(img_filtered, cmap='gray')
axes[0].set_title('Original image')
axes[1].set_title('Low-pass filtered image')
```
